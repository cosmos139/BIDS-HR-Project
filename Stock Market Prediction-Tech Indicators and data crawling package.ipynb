```python
def rsi(df,period):
    closing_price = df["Close"]
    g_l = []
    for i in range(1,len(closing_price)):
        g_l = g_l + [closing_price[i] - closing_price[i-1]]
    first_period = g_l[0:(period - 1)]
    avg_gain1 = np.mean([first_period[i] for i in range(0,len(first_period)) if first_period[i] > 0])
    avg_loss1 = abs(np.mean([first_period[i] for i in range(0,len(first_period)) if first_period[i] < 0]))
    rs1 = avg_gain1/avg_loss1
    avg_gain = [avg_gain1]
    avg_loss = [avg_loss1]
    for j in range(0,len(g_l[period:])):
        if g_l[period:][j] > 0:
            avg_gain = avg_gain + [(avg_gain[len(avg_gain)-1] * (period - 1) + g_l[period:][j])/period]
            avg_loss = avg_loss + [avg_loss[len(avg_loss)-1] * (period - 1) / period]
        elif g_l[period:][j] == 0:
            avg_gain = avg_gain + [(avg_gain[len(avg_gain)-1])*(period - 1) /period]
            avg_loss = avg_loss + [avg_loss[len(avg_loss)-1] * (period - 1) /period]
        else:
            avg_gain = avg_gain + [(avg_gain[len(avg_gain)-1])*(period - 1) /period]
            avg_loss = avg_loss + [(avg_loss[len(avg_loss)-1] * (period - 1) + abs(g_l[period:][j])) /period]
    rs = [rs1] + [avg_gain[k]/avg_loss[k] for k in range(0,len(avg_gain))]
    rsi = [100 - 100/(1+ rs[m]) for m in range(0,len(rs))]
    return rsi
```

```python
def william_pr(df,period):
    perc_r = []
    for i in range(0,len(df[period - 1:])):
        highest = max(df["High"][i:i + period])
        lowest =  min(df["Low"][i:i + period])
        close = df["Close"][i + period - 1]
        perc_r =  perc_r + [(highest - close)/(highest - lowest) * (-100)]
    return perc_r
```

```python
def ema(df,period):
    smoothing_constant = 2/(period + 1)
    first_ema = np.mean(df["Close"][0:period])
    ema_list = [first_ema]
    for i in range(0,len(df["Close"][period:])):
        ema_list = ema_list + [(df["Close"][period:][i] - ema_list[i]) * smoothing_constant + ema_list[i]]
    return ema_list
```

```python
def roc(df,period):
    roc_list = []
    for i in range(0,len(df["Close"][period:])):
        roc_list = roc_list + [(df["Close"][period:][i] - df["Close"][i])/df["Close"][i]]
    return roc_list
```

```python
def obv(df):
    obv_list = [df["Volume"][0]]
    for i in range(1,len(df["Volume"])):
        if df["Close"][i] - df["Close"][i - 1] > 0:
            obv_list = obv_list + [obv_list[i-1] + df["Volume"][i]]
        elif df["Close"][i] - df["Close"][i - 1] == 0:
            obv_list = obv_list + [obv_list[i-1]]
        else:
            obv_list = obv_list + [obv_list[i-1] - df["Volume"][i]]
    return obv_list
```

```python
def pvt(df):
    pvt_list = [0]
    for i in range(1,len(df["Close"])):
        pvt_list = pvt_list + [(df["Close"][i] - df["Close"][i - 1])/df["Close"][i - 1] * df["Volume"][i] + pvt_list[i - 1]]
    return pvt_list
```

```python
def sma(df,period):
    sma_list = []
    for i in range(0,len(df["Close"]) - period):
        sma_list = sma_list + [np.mean(df["Close"][i:(i + period)])]
    return sma_list
```

```python
import urllib.request
import datetime
import io
import requests
from pytz import timezone
import pytz
def google_finance_datareader(symbol, interval_seconds, num_days):
    url_string = "http://www.google.com/finance/getprices?q={symbol}".format(symbol=symbol.upper())
    url_string += "&i={interval_seconds}&p={num_days}d&f=d%2Cc%2Ch%2Cl%2Co%2Cv".format(interval_seconds=interval_seconds, num_days=num_days)
    page = requests.get(url_string).content
    df = pd.read_csv(io.StringIO(page.decode('utf-8')), skiprows=7, sep=',', names=['DATE', 'CLOSE', 'HIGH', 'LOW', 'OPEN', 'VOLUME'])
    b_dateround = df['DATE'].map(lambda dt: dt[0]=='a')
    dateround = df[b_dateround]['DATE'].map(lambda dt: int(dt[1:]))
    df['DATE2'] = dateround
    df['DATE2'] = df['DATE2'].fillna(method='ffill')
    df['DATE3'] = df[~b_dateround]['DATE'].astype(int)*interval_seconds
    df['DATE3'] = df['DATE3'].fillna(0)
    df['DATE4'] = df['DATE2'] + df['DATE3']
    df['DATE4'] = df['DATE4'].map(lambda s: datetime.datetime.fromtimestamp(int(s), timezone('US/Eastern')))
    del df['DATE']
    del df['DATE2']
    del df['DATE3']
    df = df.set_index('DATE4', verify_integrity=True)
    df.index.name = 'DATE'
    return(df)

symbol = 'googl'
interval_seconds = 300
num_days = 5
df = google_finance_datareader(symbol, interval_seconds, num_days)
df
```

```python
class Stock:
    def __init__(self, industry, size, stock):
        self.stock = stock
        self.size = size
        self.industry = industry
        self.dirpath = 'D:\ProStock\stocks/{0}/{1}'.format(self.industry, self.size)
        self.filepath = '{0}/{1}.csv'.format(self.dirpath, self.stock)
        self.skip = False

        # create dir if it does not exist
        try:
            os.makedirs(self.dirpath)
        except Exception as e:
            #print e
            pass

        self.start = datetime(2010, 1, 1)
        self.end = datetime(2017, 7, 19)

        try:
            self.df = pd.read_csv(self.filepath)
            if len(self.df) == 0:
                # empty DF, info not available -> skip!
                self.skip = True
        except:
            try:
                print 'Fetching', self.filepath
                self.df = pdr.get_data_google(symbols=self.stock, start=self.start, end=self.end)
                self.df.dropna(inplace=True)
                self.df.to_csv(self.filepath)
            except:
                # can't be found -- save empty file so as not to waste time afterwards
                print 'FAILED!'
                pd.DataFrame().to_csv(self.filepath)
                self.skip = True

    def indices(self, period=14):
        try:
            self.df["RSI"] = [np.nan for _ in range(period-1)] + rsi(self.df,period)
            self.df["William %R"] = [np.nan for _ in range(period-1)] + william_pr(self.df,period)
            self.df["OBV"] = obv(self.df)
        except:
            self.skip = True
```
